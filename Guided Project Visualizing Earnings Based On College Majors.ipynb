# Exploratory Data Analysis on USA Recent graduated students income between 2010 and 2012

## 1. Importing and inspectioning few rows of the data set

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Reading the data set and creating a panda Data Frame
recent_grads = pd.read_csv('recent-grads.csv')
#Visualizing the first five rows of the data set
recent_grads.head()
```

```python
recent_grads.describe()
```

```python
#Eliminating missing values
raw_data_count = len(recent_grads.index)
recent_grads = recent_grads.dropna()
# How many data were eliminated?
print('cleaned data set has',len(recent_grads.index),'rows')
print('uncleaned data set had',raw_data_count,'rows')
```

## 2. Correlating median income with other variables

The following questions are worth investigating in the data set. Such as:

1. Does people in major with higher female proportion make more money than with fewer?
2. Does people in popular majors have higher salaries?
3. Does people in full time jobs with higher incomes?

To answer then, the following scatter plots will be displayed 
* ShareWomen x Median   
* Total x Median
* Full_time_year_round x Median

```python
# using panda plots!
ax = recent_grads.plot(x='ShareWomen',y='Median',kind='scatter')
ax.set_title('ShareWomen vs Median_income_year')
```

```python
# using panda plots!
ax = recent_grads.plot(x='Total',y='Median',kind='scatter')
ax.set_title('Total vs Median_income_year')
recent_grads['Total'].max()
```

```python
# using panda plots!
ax = recent_grads.plot(x='Full_time_year_round',y='Median',kind='scatter')
ax.set_title('full_time_workers vs Median_income_year')
recent_grads['Total'].max()
```

## 3. Major Distributions through variables

This section aim to answer the following questions by using the histogram as a tool:

* What is the fraction of majors with proportion of women greater or equal to 0.5? The histogram of ShareWomen variable will be displayed as a guide for the answer
* What is the most common (mode) sallary among all the majors? The histogram of *Median* variable will be displayed

```python
ax = recent_grads['ShareWomen'].hist(bins = 10,range=(0,1))
ax1 = recent_grads['ShareWomen'].hist(bins=10,histtype='step',cumulative=True, range = (0,1))
a = np.arange(0,1+0.1,0.1).tolist()
ax.set_xticks(a)
```

It can be seen that more than 50% of the majors have more women the men recently graduated.

```python
max_income = recent_grads['Median'].max()
min_income = recent_grads['Median'].min()
interval = (max_income-min_income)/10
ax = recent_grads['Median'].hist(bins = 10,range=(min_income,max_income))
ax.set_xlim(min_income,max_income)
ax.set_xticks(np.arange(min_income,max_income+1,interval).tolist())
ax.set_xticklabels(np.arange(min_income,max_income+1,interval).round(),rotation = 40)
```

It can be seen that the most common median income is between 30800 and 39600 dolars per year.
